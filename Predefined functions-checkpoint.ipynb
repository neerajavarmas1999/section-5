```python
#56
import random
ans = random.randint(1,100)
print("Guess the hidden number between 1 and 100")
guess = 0
while ans != guess:
    guess = int(input("Enter your guess."))
    while guess<1 or guess>100:
        print("Your guess was not valid. Enter it again.")
        guess = int(input("Enter your guess."))
    if guess < ans:
        print("Your guess is too low. Try again")
    elif guess> ans:
        print("Your guess is too high. Try again")
    elif ans == guess:
        print("Correct! Well Done")
```

```python
#59
import random
total = 0
for loop in range(10):
    num1 = random.randint(1,1000)
    num2 = random.randint(1,1000)
    print("I have generated two random numbers")
    choice = int(input("Which is largest, 1 or 2?"))
    while choice<1 or choice>2:
        choice = int(input("Please enter only 1 or 2"))
    print("Number 1 =",num1)
    print("Number 2 =",num2)
    if choice == 1 and num1>num2:
        print("Correct, number 1 was the larger")
        total += 1
    elif choice == 2 and num1<num2:
        print("Correct, number 2 was the larger")
        total += 1
    else:
        print("Sorry, wrong guess")
print("Your total was",total,"correct guesses out of 10")
print("Your final score was",total,"correct")
```

```python
#60
ages = []
seats = []
another = "Y"
while another == "Y":
    age = int(input("Enter age:"))
    while age<0 or age>100:
        age = int(input("Please enter an age between 0 and 100"))
    seat = int(input("Enter seat:"))
    while seat<1 or seat>56:
        seat = int(input("Please enter a seat number between 1 and 56"))
    ages.append(age)
    seats.append(seat)
    another = str(input("Do you wish to enter another:"))
    while not(another == "Y" or another == "N"):
        another = str(input("Please enter only Y or N"))
maxage = max(ages)
priority = ages.index(maxage)
print("The oldest audience member is sitting in seat",seats[priority])
print(ages.count(0),"audience members did not give their age.")
```

```python
#61
print("Please enter the judges'scores:")
scores = str(input())
print("Please enter the difficulty rating")
rating = float(input())
separate = scores.split(",")
for loop1 in range(len(separate)):
    separate[loop1] = int(separate[loop1])
for loop2 in range(2):
    separate.remove(max(separate))
    separate.remove(min(separate))
final = sum(separate) * rating
print("The three remaining judges' scores are:")
for loop3 in range(len(separate)):
    print(separate[loop3])
print("The diver's score is:",round(final,1))
```

```python
#62
import random
dice = []
count = 0
while not(len(dice)==6):
    count += 1
    roll = random.randint(1,6)
    print("Dice rolled =",roll)
    result = False
    for loop in range(len(dice)):
        if dice[loop] == roll:
            result = True
    if result:
        dice.remove(roll)
    else:
        dice.append(roll)
    print(dice)
    print("After roll",str(count)+":",dice)
print("Game completed in",count,"rolls of the dice.")
```
